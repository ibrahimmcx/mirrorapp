```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
im =np.array(Image.open('empire.jpg'))
plt.imshow(im)
x=[200,300,400,500]
y=[200,300,400,500]
plt.plot(x,y,'-go')
plt.title('ibrahim:"empire.jpg"')

plt.show()
```

```python
from PIL import Image ,ImageFilter
im =Image.open('empire.jpg')
im_blur=im.filter(ImageFilter.GaussianBlur(radius=5))
im_blur.show()
```

```python
from PIL import ImageEnhance
im =Image.open('empire.jpg')
enhancer = ImageEnhance.Brightness(im)
im_brightened = enhancer.enhance(0.5)
im_brightened.show()
```
