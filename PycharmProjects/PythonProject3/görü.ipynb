```python
from PIL import Image
from pylab import *
# read image to array
im = array(Image.open('empire.jpg').convert('L'))
im2=255-im
im3=(100.0/255)*im+100
im4=255.0*(im/255.0)**2

fig, axs = plt.subplots(2, 2, figsize=(9, 7))

axs[0, 0].imshow(im,  cmap='gray', vmin=0, vmax=255); axs[0, 0].set_title('im (orijinal)')
axs[0, 1].imshow(im2, cmap='gray', vmin=0, vmax=255); axs[0, 1].set_title('im2 (invert)')
axs[1, 0].imshow(im3, cmap='gray', vmin=0, vmax=255); axs[1, 0].set_title('im3 (100..200 clamp)')
axs[1, 1].imshow(im4, cmap='gray', vmin=0, vmax=255); axs[1, 1].set_title('im4 (square)')

for ax in axs.ravel():
    ax.axis('off')

plt.tight_layout()
plt.show()
```
